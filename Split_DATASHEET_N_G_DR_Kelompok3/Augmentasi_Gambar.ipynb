# Augmentasi



```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
```

```python
def flip_image(image, flip_code):
    return cv2.flip(image, flip_code)

def rotate_image(image, angle):
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated_image

def shift_image(image, shift_x, shift_y):
    translation_matrix = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
    shifted_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
    return shifted_image
```

```python
# Load image
image_path = '/content/Glaucoma_(40).jpg'
image = cv2.imread(image_path)
#flip)
flipped_image = flip_image(image, 1)
#  rotasi
rotated_image_270 = rotate_image(image, 270)
rotated_image_60 = rotate_image(image, 60)
rotated_image_90 = rotate_image(image, 90)
rotated_image_30 = rotate_image(image, 30)
# Shift image
shifted_image = shift_image(image, 100, 50)
```

```python
cv2_imshow(image)
cv2_imshow(flipped_image)
cv2_imshow(rotated_image_45)
cv2_imshow(rotated_image_60)
cv2_imshow(rotated_image_30)
cv2_imshow(rotated_image_90)
cv2_imshow(shifted_image)
```

# Program lain


```python
from google.colab import files
import cv2
import numpy as np
import zipfile
import os

def flip_image(image, flip_code):
    return cv2.flip(image, flip_code)

def rotate_image(image, angle):
    height, width = image.shape[:2]
    rotation_matrix = cv2.getRotationMatrix2D((width / 2, height / 2), angle, 1)
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    return rotated_image

def shift_image(image, shift_x, shift_y):
    translation_matrix = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
    shifted_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
    return shifted_image

# Upload gambar-gambar
uploaded = files.upload()

# Load gambar-gambar
images = []
for filename in uploaded.keys():
    image = cv2.imdecode(np.frombuffer(uploaded[filename], np.uint8), -1)
    images.append(image)

# Flip image (contoh: horizontal flip)
flipped_images = [flip_image(image, 1) for image in images]

# Rotate image (contoh: rotasi 45 derajat)
rotated_images_45 = [rotate_image(image, 45) for image in images]
rotated_images_60 = [rotate_image(image, 60) for image in images]
rotated_images_90 = [rotate_image(image, 90) for image in images]
rotated_images_30 = [rotate_image(image, 30) for image in images]

# Shift image (contoh: pergeseran 100 pixel ke kanan dan 50 pixel ke bawah)
shifted_images = [shift_image(image, 100, 50) for image in images]

#Simpan gambar-gambar hasil augmentasi ke dalam file .zip
zip_path = '/content/Dataset Tambahan Normal.zip'  # Ganti dengan path file .zip yang diinginkan
with zipfile.ZipFile(zip_path, 'w') as archive:
    for i, filename in enumerate(uploaded.keys()):
        cv2.imwrite(f'/content/{filename}_flipped.jpg', flipped_images[i])  # Simpan gambar yang di-flip
        archive.write(f'/content/{filename}_flipped.jpg', arcname=f'{filename}_flipped.jpg')

        cv2.imwrite(f'/content/{filename}_rotated_45.jpg', rotated_images_45[i])  # Simpan gambar yang dirotasi 45 derajat
        archive.write(f'/content/{filename}_rotated_45.jpg', arcname=f'{filename}_rotated_45.jpg')

        cv2.imwrite(f'/content/{filename}_rotated_60.jpg', rotated_images_60[i])  # Simpan gambar yang dirotasi 60 derajat
        archive.write(f'/content/{filename}_rotated_60.jpg', arcname=f'{filename}_rotated_60.jpg')

        cv2.imwrite(f'/content/{filename}_rotated_90.jpg', rotated_images_90[i])  # Simpan gambar yang dirotasi 90 derajat
        archive.write(f'/content/{filename}_rotated_90.jpg', arcname=f'{filename}_rotated_90.jpg')

        cv2.imwrite(f'/content/{filename}_rotated_30.jpg', rotated_images_30[i])  # Simpan gambar yang dirotasi 30 derajat
        archive.write(f'/content/{filename}_rotated_30.jpg', arcname=f'{filename}_rotated_30.jpg')

        cv2.imwrite(f'/content/{filename}_shifted.jpg', shifted_images[i])  # Simpan gambar yang digeser
        archive.write(f'/content/{filename}_shifted.jpg', arcname=f'{filename}_shifted.jpg')

print(f'File {zip_path} berhasil disimpan.')
```
